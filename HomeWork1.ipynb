### Задание №1 ###

```python
data_movies = pd.read_csv(r"C:\Users\user\Downloads\ml-25m\ml-25m\movies.csv")
data_ratings = pd.read_csv(r"C:\Users\user\Downloads\ml-25m\ml-25m\ratings.csv")
```

```python
data_merge = pd.merge(data_movies,data_ratings, how ='right', on = 'movieId' )
```

```python
data_merge_filter = data_merge[data_merge.rating == 5.0]
```

```python
df = data_merge_filter.groupby(['title'] ).aggregate({'rating' : 'count'}).sort_values(by = 'rating', ascending=False).head()
```
