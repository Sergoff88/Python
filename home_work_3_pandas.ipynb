### Задание_3 ###

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значениями столбца timestamp для данного значения userId.

```python
df_ratings = pd.read_csv(r"C:\Users\Sergofan\Downloads\доп_материалы_к_пандас_1_и_2\ml-latest-small\ratings.csv")
```

```python
df_1 = df_ratings.groupby(["userId"]).agg({"rating" : "count"}).query("rating > 100")
```

```python
df_2 = df_ratings.groupby("userId")["timestamp"].apply(lambda x: x.max() - x.min())
```

```python
df_1.merge(df_2, how = "left", on = "userId")
```
